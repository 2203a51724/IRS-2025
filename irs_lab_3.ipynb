```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corpus = [
    "I love programming in Python.",
    "Python is an amazing programming language.",
    "Machine learning is a fascinating field.",
    "I love learning new things about Python and data science."
]
```

```python
# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus into a term-document matrix
X = vectorizer.fit_transform(corpus)

# Print the vocabulary (feature names)
print("Vocabulary:\n", vectorizer.get_feature_names_out())

# Display the term-document matrix as a dense array
print("\nTerm-Document Matrix:\n", X.toarray())
```

```python
# Convert to DataFrame
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Display the DataFrame
print("\nTerm-Document Matrix as DataFrame:\n", df)
```

```python
# Sum word occurrences across all documents
word_frequencies = df.sum(axis=0)

# Sort word frequencies
sorted_word_freq = word_frequencies.sort_values(ascending=False)

print("\nWord Frequencies:\n", sorted_word_freq)
```

```python
# Plot word frequencies
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_word_freq.values, y=sorted_word_freq.index, palette="viridis")
plt.title("Word Frequencies in Corpus")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.show()
```
